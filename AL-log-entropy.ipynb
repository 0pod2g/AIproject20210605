```python
df = df.drop(['Evaporation', 'Sunshine', 'Cloud9am', 'Cloud3pm'], axis = 1) ##Delete the columns which have over 30% null values##
df.info()
```

```python
df_num = df.select_dtypes('float64')
df_obj = df.select_dtypes('object')   
```

```python
numeric_null = df_num.isnull().sum() 
numeric_null
```

```python
from sklearn.impute import SimpleImputer ##Dealing with missing values##
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'median') 
df_num = pd.DataFrame(imputer.fit_transform(df_num), columns = df_num.columns)
obj_imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
df_obj = pd.DataFrame(obj_imputer.fit_transform(df_obj), columns = df_obj.columns)
df_new = pd.concat([df_obj, df_num], axis = 1, join = 'inner') 
df_new
```

```python
df_new.info()
```

```python
df_num = df_new.select_dtypes('float64')
df_obj = df_new.select_dtypes('object')
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
data_std = scaler.fit_transform(df_num) 
data_std
```

```python
cov_matrix = np.cov(data_std.T) 
cov_matrix
```

```python
eig_vals, eig_vec = np.linalg.eig(cov_matrix)

print(eig_vals)  
print('')
print(eig_vec)
```

```python
total = sum(eig_vals)
var_exp = [(i/total)*100 for i in sorted(eig_vals, reverse = True)]
var_exp
```

```python
cum_var_exp = np.cumsum(var_exp)
cum_var_exp
```

```python
per_var = np.round(cum_var_exp, decimals = 1) 
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]

plt.figure(figsize = (12, 6))
sns.lineplot(y = var_exp, x = range(1, len(var_exp)+1), marker = 'o')
plt.ylabel('Percentage of Explained Varience')
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.savefig('C:\\Users\\pod2g\\Desktop\\AI 6.5\\PCA.png')
plt.show()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 5, random_state = 123)
df_pca = pca.fit_transform(data_std)
df_pca
```

```python
df_pca.transpose()
```

```python
pca.explained_variance_ratio_
```

```python
data_std = pd.DataFrame(data_std, columns = list(df_num))
data_std
```

```python
comp = pca.components_
data_loading = pd.DataFrame(comp, columns = list(data_std))
data_loading
```

```python
from matplotlib.patches import Rectangle
fig, ax = plt.subplots(figsize = (22, 10), facecolor='w', edgecolor = 'k')
ax = sns.heatmap(data_loading, annot = True, fmt = '.2g', cbar = False, ax = ax, vmax = 1.0, vmin = 0.0, cmap = 'Blues', yticklabels = ['PC0', 'PC1', 'PC2', 'PC3', 'PC4'])

columns_max = data_loading.abs().idxmax(axis = 0)

for col, variable in enumerate(data_loading.columns):
    position = data_loading.index.get_loc(columns_max[variable])
    ax.add_patch(Rectangle((col, position), 1,1, fill = False, edgecolor = 'red', lw=3))
plt.show()
```

```python
df_obj['Location'] = df['Location'].astype('category')
df_obj['WindGustDir'] = df['WindGustDir'].astype('category')
df_obj['WindDir9am'] = df['WindDir9am'].astype('category')
df_obj['WindDir3pm'] = df['WindDir3pm'].astype('category')
```

```python
df_obj['Location'] = df_obj['Location'].cat.codes
df_obj['WindGustDir'] = df_obj['WindGustDir'].cat.codes
df_obj['WindDir9am'] = df_obj['WindDir9am'].cat.codes
df_obj['WindDir3pm'] = df_obj['WindDir3pm'].cat.codes
```

```python
df_obj['RainToday'] = pd.get_dummies(df_obj['RainToday'], drop_first = True)
df_obj['RainTomorrow'] = pd.get_dummies(df_obj['RainTomorrow'], drop_first = True)
```

```python
df_ml = pd.concat([df_pca, df_obj], axis = 1, join = 'inner')
df_ml.info()
```

```python
df_ml
```

```python
x = df_ml.drop('RainTomorrow', axis = 1)
x = np.array(x,type(int))
x
```

```python
y = df_ml.pop('RainTomorrow')
y = np.array(y,type(int))
y
```

```python
rand = np.random.RandomState(0)
indices = np.arange(len(df_ml))
rand.shuffle(indices)
```

```python
x1 = x[indices[:10390]]
y1 = y[indices[:10390]]
```

```python
x1=x1.astype('float')
y1=y1.astype('float')
```

```python
x1
```

```python
y1
```

```python
x_component, y_component = x1[:, 0], x1[:, 1] ##Testing Set##
```

```python
x_component
```

```python
y_component
```

```python
n_labeled_examples = x1.shape[0]
training_indices = np.random.randint(low=0, high=n_labeled_examples + 1, size=10)

X_train = x1[training_indices]
y_train = y1[training_indices]

X_pool = np.delete(x1, training_indices, axis=0)
y_pool = np.delete(y1, training_indices, axis=0)
```

```python
from sklearn.linear_model import LogisticRegression
from modAL.models import ActiveLearner
from modAL.uncertainty import entropy_sampling

logreg = LogisticRegression(solver = 'liblinear')
learner = ActiveLearner(estimator=logreg,query_strategy=entropy_sampling, X_training=X_train, y_training=y_train)
```

```python
predictions = learner.predict(x1)
is_correct = (predictions == y1)

predictions
```

```python
unqueried_score = learner.score(x1, y1)

fig, ax = plt.subplots(figsize=(8.5, 6), dpi=130)
ax.scatter(x=x_component[is_correct],  y=y_component[is_correct],  c='g', marker='+', label='Correct',   alpha=8/10)
ax.scatter(x=x_component[~is_correct], y=y_component[~is_correct], c='r', marker='x', label='Incorrect', alpha=8/10)
ax.legend(loc='lower right')
ax.set_title("Testing Set Class Predictions (Accuracy: {score:.3f})".format(score=unqueried_score))
plt.savefig('C:\\Users\\pod2g\\Desktop\\AI 6.5\\TestingSet1.png')
plt.show()
```

```python
unqueried_score
```

```python
N_QUERIES = 200
performance_history = [unqueried_score]

for index in range(N_QUERIES):
    query_index, query_instance = learner.query(X_pool)
    
    X,y = X_pool[query_index].reshape(1,-1), y_pool[query_index].reshape(1, )
    learner.teach(X = X, y = y)
    
    X_pool, y_pool = np.delete(X_pool, query_index, axis = 0), np.delete(y_pool, query_index)
    model_accuracy = learner.score(x1, y1)
    print('Accuracy after query {n}: {acc:0.4f}'.format(n=index + 1, acc=model_accuracy))
    
    performance_history.append(model_accuracy)
```

```python
fig, ax = plt.subplots(figsize=(8.5, 6), dpi=130)

ax.plot(performance_history)
ax.scatter(range(len(performance_history)), performance_history, s=13)

ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(nbins=5, integer=True))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(nbins=10))
ax.yaxis.set_major_formatter(mpl.ticker.PercentFormatter(xmax=1))

ax.set_ylim(bottom=0.5, top=1)
ax.grid(True)

ax.set_title('Incremental Classification Accuracy: Active Learning')
ax.set_xlabel('Query iteration')
ax.set_ylabel('Classification Accuracy')
plt.savefig('C:\\Users\\pod2g\\Desktop\\AI 6.5\\AL.png')

plt.show()
```
